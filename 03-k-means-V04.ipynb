## Load Data: 

```python
import os 
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

file='adresse.csv'
df=DataFrame(pd.read_csv(file,sep=',',index_col=0))
print(df)
nord=df.iloc[:,0]
est=df.iloc[:,1]
```

```python
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(df.iloc[:,0],df.iloc[:,1])
```

## K-means: 

```python
from sklearn.cluster import KMeans
# initializing KMeans: Estimators:
#estimators=[('kmeans-2',KMeans(n_clusters=2)), ('k-means3',KMeans(n_clusters=3))]

fignum=1
titles=['2 clusters','3 clusters']
sse={}
for k in range(1,10):
    # fitting with inputs: 
    est=KMeans(n_clusters=k)
    est.fit(df) 
    # Predicting the clusters: 
    labels=est.predict(df)
    # Getting cluster centers
    C=est.cluster_centers_
    sse[k]=est.inertia_
    fig=plt.figure(fignum,figsize=(16,9))
    ax=Axes3D(fig)
    ax.scatter(df.iloc[:,0],df.iloc[:,1],c=labels.astype(np.float),s=500)
    ax.scatter(C[:,0],C[:,1],marker='*',c='b', s=500)
    ax.set_title(str(k) + str(" clusters"))
    ax.dist = 12
    fignum += 1
```

```python
plt.figure()
plt.plot(list(sse.keys()),list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```
