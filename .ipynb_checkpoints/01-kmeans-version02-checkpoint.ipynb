# Clustering_Exo01

## Load Data: 

```python
import os 
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

file='adresse.csv'
df=DataFrame(pd.read_csv(file,sep=',',index_col=0))
print(df)
nord=df.iloc[:,0]
est=df.iloc[:,1]
```

```python
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(df.iloc[:,0],df.iloc[:,1])
```

## K-means: 

```python
from sklearn.cluster import KMeans
# initializing KMeans:
kmeans=KMeans(n_clusters=3)
# fitting with inputs: 
kmeans = kmeans.fit(df) 
# Predicting the clusters: 
labels=kmeans.predict(df)
# Getting cluster centers
C=kmeans.cluster_centers_
print(C)
```

```python
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(df.iloc[:,0],df.iloc[:,1],c=labels.astype(np.float))
ax.scatter(C[:,0],C[:,1],marker='*',c='b', s=100)
```

```python
plt.scatter(nord,est,c=labels.astype(np.float))
plt.xlabel("nord",fontsize=16)
plt.ylabel("Est",fontsize=16)
plt.scatter(C[:,0],C[:,1],marker='*',c='b', s=100)
plt.show()
```
